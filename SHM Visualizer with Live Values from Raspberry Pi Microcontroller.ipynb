```python
import json
import os
import sys
import serial
import numpy as np
from PyQt5 import QtWidgets, QtCore
import pyqtgraph as pg
from datetime import datetime


def moving_average(data, window_size=10):
    if len(data) < window_size:
        return np.array(data)
    return np.convolve(data, np.ones(window_size)/window_size, mode="valid")

class LivePlot(QtWidgets.QMainWindow):
    def __init__(self, port="COM5", baudrate=115200):
        super().__init__()

        self.ser = serial.Serial(port, baudrate, timeout=1)

        self.time_values, self.x_values, self.v_values, self.a_values = [], [], [], []

        self.bar_N = 50
        self.v_bar_buffer = []
        self.a_bar_buffer = []

        self.setWindowTitle("SHM Live Analysis")
        cw = QtWidgets.QWidget()
        self.setCentralWidget(cw)
        main_grid = QtWidgets.QGridLayout()
        cw.setLayout(main_grid)

        pg.setConfigOptions(antialias=True, background="k", foreground="w")

        self.plot_grid = QtWidgets.QGridLayout()
        main_grid.addLayout(self.plot_grid, 0, 0)

        # Plot 1: x & v
        self.plot1 = pg.PlotWidget(title="Displacement & Velocity vs Time")
        self.plot1.addLegend(offset=(20,20))
        self.plot1.setLabel('bottom', 'Time', units='s')    # x-axis
        self.plot1.setLabel('left', 'Displacement / Velocity', units='m or m/s')  # y-axis
        self.curve_x = self.plot1.plot(pen="y", name="x(t)")
        self.curve_v = self.plot1.plot(pen="c", name="v(t)")
        self.plot_grid.addWidget(self.plot1, 0, 0)

        # Plot 2: KE & PE
        self.plot2 = pg.PlotWidget(title="Energy Exchange")
        self.plot2.addLegend(offset=(20,20))
        self.curve_ke = self.plot2.plot(pen="r", name="KE")
        self.curve_pe = self.plot2.plot(pen="g", name="PE")
        self.curve_te = self.plot2.plot(pen="y", name="TE")
        self.plot_grid.addWidget(self.plot2, 0, 1)

        # Plot 3: Moving bar |v| & |a|
        self.plot3 = pg.PlotWidget(title="|Velocity| & |Acceleration| over Time")
        self.plot3.addLegend(offset=(20,20))
        self.bar_v = pg.BarGraphItem(x=[], height=[], width=0.3, brush="c")
        self.bar_a = pg.BarGraphItem(x=[], height=[], width=0.3, brush="m")
        self.plot3.addItem(self.bar_v)
        self.plot3.addItem(self.bar_a)
        self.plot_grid.addWidget(self.plot3, 1, 0)

        # Plot 4: Phase portrait
        self.plot4 = pg.PlotWidget(title="Phase Portrait (x vs v)")
        self.plot4.addLegend(offset=(20,20))
        self.curve_phase = self.plot4.plot(pen="w")
        self.plot_grid.addWidget(self.plot4, 1, 1)

        # Sidebar as a proper QWidget
        self.sidebar_widget = QtWidgets.QWidget()
        self.sidebar_widget.setFixedWidth(350)
        self.sidebar_widget.setStyleSheet("background-color: black;")
        main_grid.addWidget(self.sidebar_widget, 0, 1)  # add directly to grid

        self.sidebar_inner = QtWidgets.QVBoxLayout()
        self.sidebar_inner.setContentsMargins(15, 15, 15, 15)  
        self.sidebar_inner.setSpacing(15)
        self.sidebar_widget.setLayout(self.sidebar_inner)

        self.labels = {}
        
        # Heading
        heading = QtWidgets.QLabel("Live Parameters")
        heading.setStyleSheet("color: white; font-size: 16pt; font-weight: bold;")
        self.sidebar_inner.addWidget(heading)

        self.label_names = [
            "Displacement", "Velocity", "Acceleration",
            "Phase Angle", "Restoring Force",
            "Time Period", "Amplitude", "Frequency", "Angular Frequency",
            "KE", "PE", "TE", "Energy Dissipation Rate",
            "Max PE", "Max KE"
        ]
        for name in self.label_names:
            lbl = QtWidgets.QLabel(f"{name}: 0.00")
            lbl.setStyleSheet("color: white; font: 11pt;") 
            self.sidebar_inner.addWidget(lbl)
            self.labels[name] = lbl

        # Start button
        self.start_button = QtWidgets.QPushButton("Start")
        self.start_button.setStyleSheet(
            "background-color: #28a745; color: white; font-size: 14px; font-weight: bold; padding: 8px; border-radius: 6px;"
        )
        self.start_button.clicked.connect(self.start_data_acquisition)

        # Screenshot button
        self.screenshot_button = QtWidgets.QPushButton("Screenshot")
        self.screenshot_button.setStyleSheet("background-color: #007bff; color: white; font-size: 14px; font-weight: bold; padding: 8px; border-radius: 6px;")
        self.screenshot_button.clicked.connect(self.take_screenshot)
        
        controls_heading = QtWidgets.QLabel("Controls")
        controls_heading.setStyleSheet("color: white; font-size: 14pt; font-weight: bold;")
        settings_heading = QtWidgets.QLabel("Settings")
        settings_heading.setStyleSheet("color: white; font-size: 14pt; font-weight: bold;")
        
        self.sidebar_inner.addWidget(controls_heading)
        self.sidebar_inner.addWidget(self.start_button)
        self.sidebar_inner.addWidget(self.screenshot_button)
        self.sidebar_inner.addWidget(settings_heading)

        # Mass Input
        self.mass_label = QtWidgets.QLabel("Mass (kg):")
        self.mass_label.setStyleSheet("color: white; font-size: 11pt;")
        self.sidebar_inner.addWidget(self.mass_label)

        self.mass_spin = QtWidgets.QDoubleSpinBox()
        self.mass_spin.setMinimum(0.01)
        self.mass_spin.setMaximum(100.0)
        self.mass_spin.setValue(1.0)  # default mass
        self.mass_spin.setSingleStep(0.1)
        self.mass_spin.setStyleSheet("background-color: #222; color: white; font-size: 11pt; padding: 4px; border-radius: 4px;")
        self.sidebar_inner.addWidget(self.mass_spin)

        self.points_label = QtWidgets.QLabel("Number of Points:")
        self.points_label.setStyleSheet("color: white; font-size: 11pt;")
        self.sidebar_inner.addWidget(self.points_label)

        self.points_spin = QtWidgets.QSpinBox()
        self.points_spin.setMinimum(50)
        self.points_spin.setMaximum(2000)
        self.points_spin.setValue(350)  # default number of points
        self.points_spin.setSingleStep(50)
        self.points_spin.setStyleSheet("background-color: #222; color: white; font-size: 11pt; padding: 4px; border-radius: 4px;")
        self.sidebar_inner.addWidget(self.points_spin)
        self.points_spin.valueChanged.connect(lambda val: setattr(self, "max_points", val))
        self.max_points = self.points_spin.value()

        self.sidebar_inner.addStretch()

        self.timer = QtCore.QTimer()
        self.timer.timeout.connect(self.update)

        self.sidebar_timer = QtCore.QTimer()
        self.sidebar_timer.timeout.connect(self.update_sidebar)
        self.sidebar_timer.start(50)
            
        self.last_values = {}


    # Start/Pause functionality
    def start_data_acquisition(self):
        if self.timer.isActive():
            self.timer.stop()
            self.start_button.setText("Start")
            self.start_button.setStyleSheet(
                "background-color: #28a745; color: white; font-size: 14px; font-weight: bold; padding: 8px; border-radius: 6px;"
            )
        else:
            self.timer.start(20)
            self.start_button.setText("Pause")
            self.start_button.setStyleSheet(
                "background-color: #dc3545; color: white; font-size: 14px; font-weight: bold; padding: 8px; border-radius: 6px;"
            )

    # Screenshot functionality
    def take_screenshot(self):
        """Take screenshot of the application"""
        filename, _ = QtWidgets.QFileDialog.getSaveFileName(self, "Save Screenshot",f"shm_demo_screenshot_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png","PNG Files (*.png)")
    
        if filename:
            screen = QtWidgets.QApplication.primaryScreen()
            screenshot = screen.grabWindow(self.winId())
            screenshot.save(filename, 'png')
            QtWidgets.QMessageBox.information(self, "Success", f"Screenshot saved to:\n{filename}")


    # Data update
    def update(self):
        try:
            line = self.ser.readline().decode("utf-8").strip()
            if not line:
                return
            parts = line.split(",")
            if len(parts) < 3:
                return

            t = float(parts[0])
            x = float(parts[1])
            v = float(parts[2])

            # Compute acceleration
            if len(self.v_values) > 1:
                a = (v - self.v_values[-1]) / (t - self.time_values[-1] + 1e-6)
            else:
                a = 0.0

            self.time_values.append(t)
            self.x_values.append(x)
            self.v_values.append(v)
            self.a_values.append(a)

            if len(self.time_values) > self.max_points:
                self.time_values = self.time_values[-self.max_points:]
                self.x_values = self.x_values[-self.max_points:]
                self.v_values = self.v_values[-self.max_points:]
                self.a_values = self.a_values[-self.max_points:]

            # Smooth
            x_s = moving_average(self.x_values, 8)
            v_s = moving_average(self.v_values, 8)
            a_s = moving_average(self.a_values, 8)
            t_s = self.time_values[-len(x_s):]

            # Plot 1
            self.curve_x.setData(t_s, x_s)
            self.curve_v.setData(t_s[-len(v_s):], v_s)

            # Plot 2
            m = 1.0
            if len(x_s) > 0:
                x_max = np.max(np.abs(x_s))
                v_max = np.max(np.abs(v_s))
                k = m * (v_max / (x_max + 1e-6))**2
            else:
                k = (2*np.pi)**2 * m
            KE = 0.5*m*v_s**2
            PE = 0.5*k*x_s**2
            TE = KE + PE
            KE_s = moving_average(KE, 8)
            TE_s = moving_average(TE, 50)

            self.curve_ke.setData(t_s[-len(KE_s):], KE_s)  
            self.curve_pe.setData(t_s, PE)
            self.curve_te.setData(t_s[-len(TE_s):], TE_s)

            # Plot 3: bars
            self.v_bar_buffer.append(abs(v_s[-1]))
            self.a_bar_buffer.append(abs(a_s[-1]))
            if len(self.v_bar_buffer) > self.bar_N:
                self.v_bar_buffer = self.v_bar_buffer[-self.bar_N:]
                self.a_bar_buffer = self.a_bar_buffer[-self.bar_N:]
            indices = np.arange(len(self.v_bar_buffer))
            self.bar_v.setOpts(x=indices-0.15, height=self.v_bar_buffer, width=0.3)
            self.bar_a.setOpts(x=indices+0.15, height=self.a_bar_buffer, width=0.3)

            # Plot 4
            self.curve_phase.setData(x_s, v_s)

            self.last_values = {
                "Displacement": x_s[-1],
                "Velocity": v_s[-1],
                "Acceleration": a_s[-1],
                "Phase Angle": np.arctan2(v_s[-1], x_s[-1]),
                "Restoring Force": -k*x_s[-1],
                "Time Period": 2*np.pi/np.sqrt(k/m),
                "Amplitude": np.max(np.abs(x_s)),
                "Frequency": np.sqrt(k/m)/(2*np.pi),
                "Angular Frequency": np.sqrt(k/m),
                "KE": KE[-1],
                "PE": PE[-1],
                "TE": TE[-1],
                "Energy Dissipation Rate": (TE[-1]-TE[-2])/(t_s[-1]-t_s[-2]+1e-6) if len(TE)>1 else 0.0,
                "Max PE": np.max(PE),
                "Max KE": np.max(KE)
            }

        except Exception as e:
            print("Read error:", e)

    # --- Sidebar update ---
    def update_sidebar(self):
        for key, val in self.last_values.items():
            if key in self.labels:
                self.labels[key].setText(f"{key}: {val:.6f}")  # 6 decimals

if __name__ == "__main__":
    app = QtWidgets.QApplication(sys.argv)
    win = LivePlot(port="COM5", baudrate=115200)
    win.show()
    sys.exit(app.exec_())
```
